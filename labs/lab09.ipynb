**20) [M] In the context of association rules, explain the difference between support, confidence, lift, leverage and
conviction.**

*Support*: This is a measure of the frequency of an itemset in a database/dataframe <br>
*Confidence*: This is the probability that we see a consequent itemset in a transaction, given that the transaction already contains the antecedent itemset.<br>
*Lift*: This is a metric used to measure variation from statistical independence. More specifically, it is used to measure how much more often the antecedent and consequent of a rule A->C occur together than we would expect if they were statistically independent.<br>
*Leverage*: Somewhat similar to lift in the sense both can be used to determine independence, but leverage computes the difference between the observed frequency of A and C appearing together and the frequency that would be expected if A and C were independent.<br>
*Conviction*: A metric to indicate the degree to which a consequent itemset is dependent on an antecedent itemset.
