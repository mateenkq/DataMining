**8) [M] What is a boxplot? In your answer, please clearly indicate what it is, why one would use it during their
EDA phase, and whether it's good for one variable or to show relationships between multiple variables.**

Boxplots are a graphical display which elucidates distribution of the data through a minimum, a maximum, first quartile (75% of data has a larger value), median, third quartile (25 % of data has a larger value), and the median. The interquartile range are the values between the first and the third quartile and are represented by a box, while values outside this interquartile range are represented as whiskers that extend from the box to minimum or maximum. Again, it is good for one variable since we cannot represent multiple attributes of data on a box-whisker plot. However, if we placed boxplots adjacently, we can then get an estimate of how different variables are distributed

**12) [M] What is a quantile? What is a quartile? What is an Inter-quartile range (IQR)? Interpret the boxplot results
in these terms.**

A quantile is the division of a sample into equal-sized, adjacent subgroups - that is, it refers to the division of a probability distribution into areas of equal probability. A quartile is a quantile where the sample is divided into 4 sub-groups of equal size. The interquartile range is the middle 50% of values or the second and third quartiles, inclusive. In the boxplot, it would be represented by the box.

**13) [M, P] Read about the quantile() method for data frames, and use it to numerically show the 25th,
median, and 75th percentiles, and compute the IQR (Inter-quartile range) for both variables, on both data
frames. Compare and contrast.**

```python
print('Uniform distribution ...\n')
print('25th percentile is \n{}'.format(df_uniform.quantile(q=0.25)))
print('median is \n{}'.format(df_uniform.quantile(q=0.5)))
print('75th percentile is \n{}'.format(df_uniform.quantile(q=0.75)))
print('interquartile range is \n{}'.format(df_uniform.quantile(q=0.75) - df_uniform.quantile(q=0.25)))

print('\nNormal distribution ...\n')
print('25th percentile is \n{}'.format(df_normal.quantile(q=0.25)))
print('median is \n{}'.format(df_normal.quantile(q=0.5)))
print('75th percentile is \n{}'.format(df_normal.quantile(q=0.75)))
print('interquartile range is \n{}'.format(df_normal.quantile(q=0.75) - df_normal.quantile(q=0.25)))
```

Observe that the percentiles for uniform distribution have a greater difference between them, since they are relatively spread out (evenly over the [10, 90] interval. Thus the interquartile range of uniform distribution is much larger than normal distribution.

**14) [M, P] What is the standard definition of an outlier, in terms of IQR? Use the IQR to determine what the
outliers are for each variable in each dataset, if any.**

A defintion I liked on purplemath was that an outlier is anything more than 1.5 interquartile rages below the first quartile or above the first quartile.

```python
uniform_thres_u = df_uniform.quantile(0.75) + 1.5 * (df_uniform.quantile(q=0.75) - df_uniform.quantile(q=0.25))
uniform_thres_l = df_uniform.quantile(0.25) - 1.5 * (df_uniform.quantile(q=0.75) - df_uniform.quantile(q=0.25))
print('For uniform distribution, the outliers are given below')
print(pd.concat([df_uniform[df_uniform['x'] < uniform_thres_l['x']], df_uniform[df_uniform['x'] > uniform_thres_u['x']]])['x'].dropna())
print('\n')
print(pd.concat([df_uniform[df_uniform['y'] < uniform_thres_l['y']], df_uniform[df_uniform['y'] > uniform_thres_u['y']]])['y'].dropna())
print('\nFor normal distribution, the outliers are given below\n')
normal_thres_u = df_normal.quantile(0.75) + 1.5 * (df_normal.quantile(q=0.75) - df_normal.quantile(q=0.25))
normal_thres_l = df_normal.quantile(0.25) - 1.5 * (df_normal.quantile(q=0.75) - df_normal.quantile(q=0.25))
print(pd.concat([df_normal[df_normal['x'] < normal_thres_l['x']], df_normal[df_normal['x'] > normal_thres_u['x']]])['x'].dropna())
print('\n')
print(pd.concat([df_normal[df_normal['y'] < normal_thres_l['y']], df_normal[df_normal['y'] > normal_thres_u['y']]])['y'].dropna())
```
